## 1. OneHotEncoding using Pandas


```python
print(pd.get_dummies(df, columns=['fuel', 'owner'], prefix=['F', 'O']))
```

```python
print(pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True))
# we drop one input bcz of multicolinearity()
```
